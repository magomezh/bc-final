```python
import pandas as pd
```

```python
file="misti scraped data/apple.csv"
```

```python
df=pd.read_csv(file)
df.head
```

```python
df[['Stars','Rating']] =df.Rating.str.split(":", expand=True)
```

```python
df[['Stars', 'type1', 'type2']] =df.Stars.str.split("s", expand=True)
```

```python
df.head()
```

```python
df3 = df.drop(columns=['type1', 'type2'])
```

```python
df3['Stars'] = df['Stars'].astype(float)
```
